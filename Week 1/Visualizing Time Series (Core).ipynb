```python
import pandas as pd

url= "https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df= pd.read_csv(url)
df
```

```python
df['date']= pd.to_datetime(df['date'], format= "%Y%m%d")
df
```

```python
df= df.set_index('date')
df
```

```python
df= df.loc['2000':, ['precipitation','min_temp','mean_temp','max_temp', 'snow_depth']]
df
```

```python
#Q1: What month had the most precipitation between 2000 through 2010?
max_precip_month= df['precipitation'].resample('M').sum()
max_precip_month
```

```python
max_precip_month.idxmax()
```

```python
df_00_10= df.loc['2000':,'precipitation']
```

```python
import matplotlib.ticker as mticks
import matplotlib.dates as mdates

loc_year = mdates.YearLocator()
```

```python
ax = df_00_10.plot(figsize=(12,3))
ax.set(title='Precipitation for 2000-2010', ylabel='Inches')

# customize major ticks 
major_loc =  mdates.YearLocator(1)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)
minor_fmt = mdates.DateFormatter("%b")

ax.xaxis.set_minor_locator(minor_loc)

## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)
```

```python
#Q2: Which year between 2000-2020 had the coolest average temperature?

min_temp_yr= df['min_temp'].resample('Y').mean()
min_temp_date= min_temp_yr.idxmin()
```

```python
ax = min_temp_yr.plot(figsize=(12,3))
ax.set(title='Average Temperature', ylabel='Degrees')

# customize major ticks 
major_loc =  mdates.YearLocator(5)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)
minor_fmt = mdates.DateFormatter("%b")

ax.xaxis.set_minor_locator(minor_loc)

## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)
```

```python
ax= ax.axvline(x=min_temp_date, color='red', linestyle='--', label='Lowest Temp Date')
ax
```
