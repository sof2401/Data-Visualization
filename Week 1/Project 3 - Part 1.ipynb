```python
import pandas as pd

df= pd.read_csv('Data/Chicago crime data/Chicago-Crime_2022.csv')
df
```

```python
melted= pd.melt(df, id_vars= ['Date','District', 'Primary Type'])
melted
```

```python
melted['Date'] = pd.to_datetime(melted['Date'])
melted.info()
```

```python
melted= melted.set_index("Date")
melted
```

```python
df_ts = melted.groupby('District').resample("D").sum()
df_ts
```

```python
#Topic 1) Comparing Police Districts

df_ts['District'].value_counts().sort_values(ascending=True)
```

```python
#Topic 2) Crimes Across the Years

df_ts = df.groupby("District").resample("Y").sum()
df_ts.plot()
```

```python
#Topic 4) Comparing Months

#Answer the question: ​What months have the most crime? What months have the least?
#Answer the question: Are there any individual crimes that do not follow this pattern? If so, which crimes?


df_month = df.groupby("District").resample("M").sum()
df_month.plot()
```
