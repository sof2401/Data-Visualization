```python
import pandas as pd

url= "https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df= pd.read_csv(url)
df
```

```python
df['date']=pd.to_datetime(df['date'], format='%Y%m%d')
```

```python
df
```

```python
df = df.set_index('date')
df
```

```python
df= df.loc['2000':'2020']
```

```python
df
```

```python
df= df.drop(['cloud_cover', 'sunshine', 'global_radiation','pressure'], axis= 1)
df
```

```python
#a) "The lowest temperature of {max temp} degrees occurred on {date as "MonthName 2-digit-day, 4-digit year. } which was a {day of the week}."

max_temp= df['max_temp'].idxmax()
min_temp= df['min_temp'].idxmin()
print(f"The lowest temperature of {min_temp} degrees occured on {min_temp.strftime('%B %d, %Y')}")
```

```python
#b) "The highest daily precipitation of {max preciptation} inches occurred on {date as "MonthName 2-digit-day, 4-digit year. } which was a {day of the week}."

max_prec= df['precipitation'].idxmax()
min_prec= df['precipitation'].idxmin()
print(f"The highest daily preciptation was {max_prec} inches, which occured on {max_prec.strftime('%B %d, %Y')} which was a {max_prec.day_name()}.")
```
