```python
import pandas as pd


# PART 1
url= 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv'
df= pd.read_csv(url)
df
```

```python
large_cities= pd.DataFrame(df[df['SizeRank'].isin([0,1,2,3])])
large_cities
```

```python
melted= pd.melt(large_cities,id_vars=['RegionName','State','Metro','CountyName'])
melted
```

```python
melted = pd.melt(large_cities.drop(columns=['Metro','CountyName']),
                 id_vars=['RegionName', 'State'],
                value_name='Confirmed',var_name="Date")
melted
```

```python
melted = melted[~melted['Date'].isin(['RegionID', 'SizeRank','StateName','State','RegionType'])]
melted
```

```python
melted['Date'] = pd.to_datetime(melted['Date'])
melted.info()
```

```python
melted= melted.set_index("Date")
melted
```

```python
df_ts = melted.groupby('RegionName').resample("D").sum()
df_ts
```

```python
# PART 2

import matplotlib.ticker as mticks

def millions(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000_000
    return f"{new_x:,.0f}M"
def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"{new_x:,.0f}K"
# Create the formatters
fmt_mill = mticks.FuncFormatter(millions)
fmt_thous = mticks.FuncFormatter(thousands)
```

```python
chicago_ts = df_ts.loc['Chicago'].copy()
ax = chicago_ts.plot()
ax.set_title('Chicago')
ax.set_ylabel('Dollars')

ax.yaxis.set_major_formatter(fmt_mill)
ax.yaxis.set_major_formatter(fmt_thous)
```
